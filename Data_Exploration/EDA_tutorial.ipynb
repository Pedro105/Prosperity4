```python
from pathlib import Path

import matplotlib.pyplot as plt
import pandas as pd

# If the notebook kernel cwd is Data_Exploration/, this works:
DATA_DIR = Path("..") / "TUTORIAL_ROUND_1"
# Otherwise use an absolute path, e.g.:
# DATA_DIR = Path("/Users/pedropinto/Prosperity4/TUTORIAL_ROUND_1")

pd.set_option("display.max_columns", 20)
```

```python
def prep(df: pd.DataFrame) -> pd.DataFrame:
    out = df.copy()
    out["timestamp"] = pd.to_numeric(out["timestamp"], errors="coerce")
    out["price"] = pd.to_numeric(out["price"], errors="coerce")
    out["quantity"] = pd.to_numeric(out["quantity"], errors="coerce")
    return out

d1 = prep(load_trades(DATA_DIR / "trades_round_0_day_-1.csv", "day_-1"))
d2 = prep(load_trades(DATA_DIR / "trades_round_0_day_-2.csv", "day_-2"))

# Per-day views only
display(d1.head(10))
display(d2.head(10))
display(d1.groupby("symbol").size())
display(d2.groupby("symbol").size())
```

```python
for day_name, df in [("day_-1", d1), ("day_-2", d2)]:
    symbols = sorted(df["symbol"].dropna().unique())
    fig, axes = plt.subplots(len(symbols), 1, figsize=(12, 3.5 * len(symbols)))
    if len(symbols) == 1:
        axes = [axes]
    for ax, sym in zip(axes, symbols):
        g = df[df["symbol"] == sym].sort_values("timestamp")
        ax.plot(g["timestamp"], g["price"], marker="o", ms=2, lw=1)
        ax.set_title(f"{day_name} — {sym}")
        ax.set_xlabel("timestamp")
        ax.set_ylabel("price")
        ax.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()
```
